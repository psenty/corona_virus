Let's try to investigate the time evolution

```python
total_cases = [] # list of confirmed cases
dates = [] # list of dates

for i in range(22,32): # in January
    date1 = "01/%.2d/20" % (i,) # first possible date format
    date2 = "01/%.2d/2020" % (i,) # second possible date format
    dates+=["01/%.2d" % (i,)] 
    total_number_cases = (np.nansum(data[data.ObservationDate==date1]["Confirmed"].values)  
        + np.nansum(data[data.ObservationDate==date2]["Confirmed"].values))
    total_cases.append(total_number_cases)

for i in range(1,29): # in February
    date1 = "02/%.2d/20" % (i,) # first possible date format
    date2 = "02/%.2d/2020" % (i,) # second possible date format
    dates+=["02/%.2d" % (i,)] 
    total_number_cases = (np.nansum(data[data.ObservationDate==date1]["Confirmed"].values)  
        + np.nansum(data[data.ObservationDate==date2]["Confirmed"].values))
    total_cases.append(total_number_cases)
    
for i in range(1,12): # in March
    date1 = "03/%.2d/20" % (i,) # first possible date format
    date2 = "03/%.2d/2020" % (i,) # second possible date format
    dates+=["03/%.2d" % (i,)] 
    total_number_cases = (np.nansum(data[data.ObservationDate==date1]["Confirmed"].values)  
        + np.nansum(data[data.ObservationDate==date2]["Confirmed"].values))
    total_cases.append(total_number_cases)
    
```

Bug because of 29/02, I remove it from the data

```python
import datetime as dt
import matplotlib.dates as mdates

x = [dt.datetime.strptime(d,"%m/%d").date() for d in dates]
```

```python
import matplotlib.dates as mdates

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=5))
plt.plot(x,total_cases)
plt.gcf().autofmt_xdate()

plt.ylabel("Confirmed cases worldwide")
```

Logarithmic scale

```python
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=5))
plt.plot(x,total_cases)
plt.yscale('symlog')
plt.gcf().autofmt_xdate()

plt.ylabel("Confirmed cases worldwide")
```
